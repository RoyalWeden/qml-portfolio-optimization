```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/all_stocks_5yr.csv')
df
```

```python
df = df.dropna()
df
```

```python
df = df.rename(columns={'Name': 'symbol'})
df
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
date_enc = LabelEncoder()
date_enc.fit(df.loc[:, 'date'])
symbol_enc = LabelEncoder()
symbol_enc.fit(df.loc[:, 'symbol'])
```

```python
date_encoded = pd.DataFrame(date_enc.transform(df.loc[:, 'date']), columns=['date'])
date_encoded
```

```python
symbol_encoded = pd.DataFrame(symbol_enc.transform(df.loc[:, 'symbol']), columns=['symbol'])
symbol_encoded
```

```python
scaled_volume = df['volume'].apply(lambda x: x/1000000) # convert volume to millions
scaled_volume
```

```python
oc_change_by_volume = pd.DataFrame((df['close'] - df['open']) * scaled_volume, columns=['oc_change_by_volume'])
oc_change_by_volume
```

```python
df = symbol_encoded.merge(date_encoded, left_index=True, right_index=True)\
                   .merge(oc_change_by_volume, left_index=True, right_index=True)
df
```

```python
mean_df = pd.DataFrame(columns=['symbol', 'mean_occbv'])
mean_df
```

```python
for i in range(max(df['symbol'])):
    selected_vals = df.loc[df['symbol']==i]['oc_change_by_volume']
    mean_df = mean_df.append({'symbol': i, 'mean_occbv': np.mean(selected_vals)}, ignore_index=True)
mean_df
```

```python
from qiskit import Aer, QuantumCircuit
from qiskit.opflow import Z, I, StateFn
from qiskit.utils import QuantumInstance
from qiskit.circuit import Parameter
from qiskit.circuit.library import RealAmplitudes, ZZFeatureMap
from qiskit.algorithms.optimizers import COBYLA, L_BFGS_B

from qiskit_machine_learning.neural_networks import TwoLayerQNN, CircuitQNN
from qiskit_machine_learning.algorithms.regressors import NeuralNetworkRegressor, VQR

from typing import Union

from qiskit_machine_learning.exceptions import QiskitMachineLearningError
```

```python
quantum_instance = QuantumInstance(Aer.get_backend('aer_simulator'), shots=1024)
```

```python
aal_df = df.loc[df['symbol']==i].drop(columns=['symbol']).set_index('date')
aal_df
```

```python
# num_samples = 100
# eps = 0.2

# X = np.array(range(0, len(aal_df), len(aal_df)//num_samples))
# y = aal_df.iloc[:num_samples].to_numpy()

X = np.array(range(len(aal_df)))
y = aal_df.to_numpy()

# plt.plot(X_, y_, 'r--')
plt.plot(X, y, 'bo')
plt.show()
```

```python
# create simple feature map
param_x = Parameter('x')
feature_map = QuantumCircuit(1, name='fm')
feature_map.ry(param_x, 0)

# create simple ansatz
param_y = Parameter('y')
ansatz = QuantumCircuit(1, name='vf')
ansatz.ry(param_y, 0)

# create QNN
regression_opflow_qnn = TwoLayerQNN(1, feature_map, ansatz, quantum_instance=quantum_instance)
```

```python
# create regressor from neural network
regressor = NeuralNetworkRegressor(neural_network=regression_opflow_qnn, loss='l2', optimizer=L_BFGS_B())
```

```python
X = X.reshape(len(X),1)
X.shape
```

```python
# fit to data
regressor.fit(X, y)

# score the result
regressor.score(X, y)
```

```python
# plot data
plt.plot(X, y, 'bo')

# plot fitted line
y_ = regressor.predict(X)
plt.plot(X, y_, 'g-')
plt.show()
```

```python
vqr = VQR(feature_map=feature_map,
          ansatz=ansatz,
          optimizer=L_BFGS_B(),
          quantum_instance=quantum_instance)
```

```python
# fit regressor
vqr.fit(X, y)

# score result
vqr.score(X, y)
```

```python
# plot target line
plt.plot(X[1200:], y[1200:], 'r--')

# plot data
plt.plot(X[1200:], y[1200:], 'bo')

# plot fitted line
y_ = vqr.predict(X)
plt.plot(X[1200:], y_[1200:], 'g-')
plt.show()
```
